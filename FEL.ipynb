```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'FEL.xlsx'  # Update the path if needed
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Plotting the bar chart
plt.figure(figsize=(10, 8))
plt.bar(df['Mutation_type'], df['Gibbs free energy (kJ/mol)'], color='skyblue')

# Add titles and labels
plt.title('Gibbs Free Energy by Mutation Type in HER2 Kinase Domain', fontsize=14)
plt.xlabel('Mutation Type', fontsize=10)
plt.ylabel('Gibbs Free Energy (kJ/mol)', fontsize=12)
plt.xticks(rotation=35)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'FEL.xlsx'  # Replace with correct path if needed
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Plotting the bar chart using categorical data for correct alignment
plt.figure(figsize=(14, 6))
plt.bar(df['Mutation_type'].astype(str), df['Gibbs free energy (kJ/mol)'], color='skyblue')

# Improve readability of x-axis labels
plt.xticks(rotation=90, ha='center')

# Add titles and labels
plt.title('Gibbs Free Energy by Mutation Type in HER2 Kinase Domain', fontsize=16)
plt.xlabel('Mutation Type', fontsize=12)
plt.ylabel('Gibbs Free Energy (kJ/mol)', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

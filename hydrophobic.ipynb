```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data
df = pd.read_excel("Hydrophobic_plot.xlsx")  # Adjust path if needed

# Set 'Mutation_type' as index for the heatmap
heat_data = df.set_index('Mutation_type')[['Area', 'Number of contacts', 'Contacts/Residue', 'Area/Residue']]

# Create the heatmap
plt.figure(figsize=(12, 14))
sns.heatmap(heat_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5, linecolor='gray')

# Customize plot
plt.title("Heatmap of Hydrophobic Interaction Metrics Across Variants", fontsize=16)
plt.xlabel("Metric", fontsize=12)
plt.ylabel("Mutation Type", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

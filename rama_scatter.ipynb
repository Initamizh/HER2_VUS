```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text  # Install this via: pip install adjustText

# Load your data
df = pd.read_excel("rama_new_plot.xlsx")  # Update the path if needed

# Set seaborn style
sns.set(style='whitegrid')

# Create the plot
plt.figure(figsize=(16, 10))
scatter = sns.scatterplot(
    x='Ramachandran favoured(%)',
    y='Ramachandran outliers(%)',
    data=df,
    hue='Ramachandran outliers(%)',
    palette='viridis',
    size='Ramachandran outliers(%)',
    sizes=(50, 300),
    edgecolor='black',
    linewidth=0.5,
    legend='brief'
)

# Add text labels with adjustText to prevent overlap
texts = [
    plt.text(
        df['Ramachandran favoured(%)'][i],
        df['Ramachandran outliers(%)'][i],
        df['Mutation_type'][i],
        fontsize=9
    )
    for i in range(df.shape[0])
]

adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

# Final formatting
plt.title('Ramachandran Favored vs Outlier Percentages (per Variant)', fontsize=18)
plt.xlabel('Favored (%)', fontsize=14)
plt.ylabel('Outlier (%)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True)
plt.legend(title='Outlier (%)', fontsize=10, title_fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'dssp_data.xlsx'  # Adjust the path if needed
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Set the index to Mutation_type for labeling
df_heatmap = df.set_index('Mutation_type')

# Set figure size
plt.figure(figsize=(12, 8))

# Create the heatmap
sns.heatmap(df_heatmap, annot=True, fmt=".2f", cmap="viridis", linewidths=0.5)

# Add titles and axis labels
plt.title("Secondary Structure Composition by Mutation Type", fontsize=16)
plt.xlabel("Structure Type", fontsize=12)
plt.ylabel("Mutation Type", fontsize=12)

# Improve layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'dssp_data.xlsx'  # Update this path if necessary
df = pd.read_excel(file_path, sheet_name='Sheet1')

# Set up the data
mutation_types = df['Mutation_type']
structure_types = df.columns[1:]  # All columns except 'Mutation_type'

# Set up the plot
fig, ax = plt.subplots(figsize=(14, 8))
bottom = [0] * len(df)

# Colors (optional: can use a colormap for more variety)
colors = plt.cm.tab20.colors

# Plot each structure type stacked on top of the previous
for i, structure in enumerate(structure_types):
    values = df[structure]
    ax.bar(mutation_types, values, bottom=bottom, label=structure, color=colors[i % len(colors)])
    bottom = [b + v for b, v in zip(bottom, values)]

# Add labels and formatting
ax.set_xlabel("Mutation Type", fontsize=12)
ax.set_ylabel("Percentage (%)", fontsize=12)
ax.set_title("Stacked Bar Chart of Secondary Structure Composition by Mutation Type", fontsize=16)
ax.legend(title="Structure Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
